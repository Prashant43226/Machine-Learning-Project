```python
ones = np.ones([X.shape[0],1])
X = np.concatenate((ones,X),axis=1)
alpha=0.005
no_of_iterations=1500
theta = np.zeros([1,5])
##cost function
def cost_function(X,theta,Y):
    product=np.dot(X,theta.T)
    residual=np.power(((X @ theta.T)-Y),2)
    cost=(np.sum(residual)/(2*len(X)))
    return cost;
cost_function(X,theta,Y)
```

```python
def gradientDescent(X,Y,theta,no_of_iterations,alpha):
    cost = np.zeros(no_of_iterations)
    for i in range(no_of_iterations):
        theta=np.zeros([1,5])
        theta=theta-(alpha/len(X))*((X.T.dot(X @ theta.T - Y)).T)
        cost=cost_function(X,theta,Y)
        return theta,cost;
gradientDescent(X,Y,theta,no_of_iterations,alpha)
```
