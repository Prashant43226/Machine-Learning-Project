### Query 6.2 
Calculate average annual percentage return and volatility of all 30 stocks over a theoretical one year period<p>

```python
pct=shares.pct_change()                     #calculate the each day percenntage change in the stocks
perc=pct.iloc[1:366,0:29]
#calculate the annual percentage return
annual_returns=pd.DataFrame(np.sum(perc)) 
annual_returns.columns=['returns']
#calculate the volatility of stocks using standard deviation method
volatility=pd.DataFrame(perc.std())        
volatility.columns=['volatility']
```

### Query 6.3
Cluster the 30 stocks according to their mean annual Volatilities and Returns using K-means clustering. Identify the optimum number of clusters using the Elbow curve method<p>

```python
data=pd.DataFrame(annual_returns)
data.columns=['returns']
data['volatility']=volatility
plt.scatter(data.iloc[:,0],data.iloc[:,1],label='True Position')     #scatter plot of all the data points
```

```python
data
```

```python
#finding the optimum number of clusters using the elbow curve method
from sklearn.cluster import KMeans 
distance=[]  #list with name distance for squared error function
k=range(1,15)
for i in k:
    km=KMeans(n_clusters=i)
    km=km.fit(data)
    distance.append(km.inertia_)
plt.plot(k,distance)
plt.ylabel('squared error cost')
plt.xlabel('K value')
##shows elbow value is 2 ie 2 clusters are the most optimum number of clusters
```

```python
from sklearn.cluster import KMeans  #import the kmeans classifier from sklearn library                                         
#cluster the datapoints using kmeans classifier
km=KMeans(n_clusters=28)            
km.fit(data)
##plot a scatter plot with different colours for different stocks
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=km.labels_, cmap='rainbow',s=50)   
plt.xlabel('returns')
plt.ylabel('volatility')
df = pd.DataFrame({'labels':km.labels_,'companies':data.index})
```
